# LDA analysis

First pass at machine learning -- unsupervised Latent Dirichlet Allocation (LDA) using the <code>gensim</code> package.

```python
import gensim
import lda
```

We'll try out the <code>pyLDAvis</code> toolkit to visualize results here too.

```python
import pyLDAvis
```

<code>lda</code> module we've written has wrappers to hook into our SQL database.  Let's pull a small dataset to work with.

```python
corpus,index = lda.build_corpus(drug=None,limit=1000)
```

<code>corpus</code> is a list of vectorized, tokenized texts -- that is, each entry in the list is a list of tuples, each tuple containing a numeric index and a frequency count.  <code>index</code> is a <code>gensim.corpora.Dictionary</code> object, a dictionary mapping numeric indices to words.  This allows us to vectorize our text corpus for efficient numerical processing.

```python
corpus
```

```python
print index
```

let's train a new LDA model on this in gensim, using the built-in multicore handler.

```python
model = gensim.models.ldamulticore.LdaMulticore(corpus=corpus,num_topics=5,id2word=index,workers=4,passes=20)
```

```python
print model
```

```python
model.show_topics()
```

so we have a vectorization of the components and weights feeding into each of our five topics.  The model is trained on our 100 input texts.  Let's try visualizing!

```python
pyLDAvis.enable_notebook()
```

```python
import pyLDAvis.gensim
pyLDAvis.gensim.prepare(model,corpus,index)
```

let's try this with another text.

```python
text = lda.stream_tokens(drug='escitalopram',limit=1)
for i in text:
    text = i
```

```python
lda_vector=model[index.doc2bow(text)]
```

```python
lda_vector
```

so it has assigned itself fairly strongly into one category.  I suspect this post was in the training set though, so this is to be expected.  The pyLDAvs method doesn't like having entries in the dict that weren't in the corpus for some reason, so we can't visualize.
