```python
def remap(s):
    if s == '\x01':
        return True
    elif s == '\x00':
        return False
    else:
        return s
```

```python
remap(u'\x01')
```

```python
remap('\x00')
```

```python
drugs['bupropion']=drugs['bupropion'].apply(lambda s: remap(s))
```

```python
for col in drugs.columns:
    drugs[col] = drugs[col].apply(lambda s: remap(s))
```

```python
subs = pandas.read_sql("select * from Subreddits",conn,index_col='subreddit')
```

```python
subs.head(10)
```

```python
subs.sort_values(by='posts',ascending=False,inplace=True)
```

```python
subs.head(10)
```

```python
fig = plt.figure() # Create matplotlib figure
ax = fig.add_subplot(111) # Create matplotlib axes
ax2 = ax.twinx() # Create another axes that shares the same x-axis as ax.
width = 0.5
subs.head(15).subscribers.plot(kind='bar', color='red', ax=ax, width=width, position=1)
subs.head(15).posts.plot(kind='bar', color='blue', ax=ax2, width=width, position=0)

ax.set_ylabel('subscribers')
ax2.set_ylabel('posts')
plt.tight_layout()
plt.savefig('/home/jrwalk/Documents/Insight/week4/subreddits.pdf')
```
