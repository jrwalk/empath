# top posts normalization

original idea: return top posts mentioning a drug based on raw score.

Problem: this biases sampling towards larger, higher-traffic subreddits.

Goal: weight score by subreddit population.

Tradeoff: simpler scrape achievable entirely within SQL query.  Weighting scheme will require pulling whole DB for calculation (should still be fast).

```python
import pandas
import pymysql as pms
import numpy
```

```python
conn = pms.connect(host='localhost',user='root',passwd='',db='empath',charset='utf8',init_command='SET NAMES UTF8')
```

```python
easy_top = pandas.read_sql("SELECT c.author,c.body,c.created_utc,c.score,c.subreddit "
                           "from Comments c ORDER BY c.score DESC LIMIT 5",conn)
```

```python
easy_top
```

all high-scoring, all from very high-traffic subreddits ([/r/AskReddit](http://www.reddit.com/r/AskReddit) and [/r/AdviceAnimals](http://www.reddit.com/r/AdviceAnimals) are among the largest overall subs).  Let's do better.

```python
better_top = pandas.read_sql("SELECT c.author,c.body,c.created_utc,c.score,c.subreddit,s.subscribers "
                             "FROM Comments c JOIN Subreddits s WHERE s.subreddit=c.subreddit "
                             "ORDER BY c.score DESC",conn)
```

```python
better_top.head(5)
```

same posts, but now we can modify the order by a normalized score.  Let's first try just norming to the population:

```python
better_top['normscore'] = pandas.Series(better_top['score']/better_top['subscribers'])
```

```python
better_top.head(10)
```

Let's try sorting this by the norm score.

```python
better_top.sort(columns='normscore',ascending=False,inplace=True)
better_top.head(10)
```

hmmm... we have some bum subreddits here.  How many of those are there (out of about 29k total posts?

```python
cur = conn.cursor()
cur.execute("SELECT subreddit FROM Subreddits WHERE subscribers<10")
```

not bad at all.  Widen it a little?

```python
cur.execute("SELECT subreddit FROM Subreddits WHERE subscribers<100")
```

Few! We've only got a few bum subreddits (unlike Dan, whose dataset is like 60% spam comments.  Turns out no one does antidepressants recreationally.  Let's just drop these infinite values.

```python
better_top = better_top[better_top.subscribers>0]
```

```python
better_top.head(10)
```

no good.  Still need to drop the tiniest subreddits.

```python
better_top = better_top[better_top.subscribers > 100]
```

```python
better_top.head(10)
```

still weak.  What if our normalization is too strong?  Let's try norming by log(N) instead.

```python
better_top.normscore = better_top.score/numpy.log10(better_top.subscribers)
```

```python
better_top.sort(columns='normscore',ascending=False,inplace=True)
better_top.head(10)
```

now our normalization is pretty weak... we're still getting largely the giant subreddits.  Let's try Poisson-like normalization, sqrt(N).

```python
better_top.normscore = better_top.score/numpy.sqrt(better_top.subscribers)
better_top.sort(columns='normscore',ascending=False,inplace=True)
better_top.head(10)
```

now we're doing quite a bit better!  getting a mix of subreddits in there now, across a range of subreddit sizes.  Think we can go with this (though we still need to do a little subreddit cleanup in the database...)
