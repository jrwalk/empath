# test TF-IDF stacking on LDA

we'll try stacking TF-IDF in <code>gensim</code> on the LDA analyzer developed before.

```python
import gensim
import lda
```

load in the visualizer too

```python
import pyLDAvis
```

```python
import warnings
warnings.simplefilter("ignore")
```

```python
corpus,index = lda.build_corpus(drug=None,limit=1000)
```

```python
tfidf = gensim.models.TfidfModel(corpus=corpus)
```

```python
lda = gensim.models.LdaMulticore(tfidf[corpus],id2word=index,passes=20,num_topics=3,workers=4)
```

```python
lda.show_topics()
```

let's see what this looks like.

```python
pyLDAvis.enable_notebook()
import pyLDAvis.gensim
pyLDAvis.gensim.prepare(lda,corpus,index)
```

hmm... this is almost entirely filtering into a single topic...
